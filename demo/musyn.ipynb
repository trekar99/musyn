### Pipe for using a Diffusion Model

```python
vae = AutoencoderKL.from_pretrained(
  'madebyollin/sdxl-vae-fp16-fix',
  use_safetensors=True,
  torch_dtype=torch.float16,
).to('cuda')
```

```python
pipe = AutoPipelineForText2Image.from_pretrained(
    "stabilityai/sdxl-turbo",
    use_safetensors=True,
    torch_dtype=torch.float16,
    variant="fp16",
    vae=vae
  )
```

```python
from diffusers import AutoencoderTiny
pipe.vae = AutoencoderTiny.from_pretrained('madebyollin/taesd', torch_device='cuda', torch_dtype=torch.float16)
pipe.vae = pipe.vae.cuda()
```

```python
pipe.to("cuda") # Indica propiedades
```

```python
#from DeepCache import DeepCacheSDHelper
#helper = DeepCacheSDHelper(pipe=pipe)
#helper.set_params(cache_interval=3, cache_branch_id=0)
#helper.enable()
```

```python
pipe.set_progress_bar_config(disable=True)
```

```python
def image_generator(prompt):
    return pipe(prompt=prompt,
                num_inference_steps=1,
                guidance_scale=0.0
                ).images[0]
```

### GUI

#### Gradio Interface GUI

```python
image_generator_demo = gr.Interface(
    image_generator,
    gr.Textbox(label="Prompt", placeholder="Enter your prompt here...", ),
    gr.Image(streaming=True),
    live=True,
    show_progress="hidden"
)

demo = gr.TabbedInterface([image_generator_demo], ["RT Image Generator"])

demo.queue().launch()
```

#### Gradio Blocks GUI

```python
with gr.Blocks() as demo:
    input = gr.Textbox(label="Input Box")
    output = gr.Image(label="Output Box")
    greet_btn = gr.Button("Greet")
    greet_btn.click(fn=image_generator, inputs=input, outputs=output, api_name="greet", show_progress='hidden')

demo.launch()
```
